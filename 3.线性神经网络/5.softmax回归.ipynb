## **复习时间 → 成绩等级 Softmax 多分类完整例子**

### **1. 场景设定：复习时间 → 成绩等级（三分类）**

我们要做一个模型：<br>
输入是：**复习时间 x（小时）**<br>
输出是：**成绩等级 y**，分 3 类：<br>
* 类别 0：不及格
* 类别 1：及格
* 类别 2：优秀<br>
目标：给定一个学生复习了 x 小时，模型输出
> 这个学生“不及格 / 及格 / 优秀”的概率各是多少？

### **2. 模型形式：一维特征 + 三个“打分” + softmax**

我们**先假设**模型已经训练好了，学出了一些参数。<br>
对每一个类别 j∈{0,1,2}，模型都会算一个打分（logit）：<br>

\begin{aligned}
z_0(x) &= -1 \cdot x + 3  && \text{(不及格的打分)} \\
z_1(x) &= 0.2 \cdot x      && \text{(及格的打分)}   \\
z_2(x) &= 1 \cdot x - 4    && \text{(优秀的打分)}
\end{aligned}

可以先看下这 3 条直线的大致趋势：
* 复习时间很少时（x 小），
    * 不及格的打分 $z_0$ 比较高 → 不及格更可能
* 复习时间中等时（x 中等），
    * 及格的打分 $z_1$ 比较高 → 及格更可能
* 复习时间很多时（x 大），
    * 优秀的打分 $z_2$ 比较高 → 优秀更可能<br>
> 这些只是“原始打分”，可以是任意实数，还不是概率。

#### **softmax 把“打分”变成“概率”**
给定一个 x，我们先算出 3 个打分$z_0, z_1, z_2$，然后用 softmax 算出概率：
$$p_j=\frac{e^{z_j}}{e^{z_0}+e^{z_1}+e^{z_2}}, j=0,1,2$$
* $p_0$：不及格的概率
* $p_1$：及格的概率
* $p_2$：优秀的概率<br>
并且三者加起来一定是 1。


### **3. 具体算几个学生：x=1 小时，3 小时，6 小时**
我们来算三个学生：
* 学生 A：复习 1 小时，真实结果：不及格（类别 0）
* 学生 B：复习 3 小时，真实结果：及格（类别 1）
* 学生 C：复习 6 小时，真实结果：优秀（类别 2）
*
#### **3.1 学生 A：只复习了 1 小时，真实“不及格”**
**第一步：算 3 个打分**
\begin{aligned}
 & x=1 \\
 & z_{0}=-1\cdot1+3=2 \\
 & z_{1}=0.2\cdot1=0.2 \\
 & z_{2}=1\cdot1-4=-3
\end{aligned}
**第二步：取指数**
\begin{aligned}
& e^{z_0}=e^2\approx7.39 \\
& e^{z_1}=e^{0.2}\approx1.22 \\
& e^{z_2}=e^{-3}\approx0.05
\end{aligned}
**第三步：求和（做分母）**
\begin{aligned}
& S=7.39+1.22+0.05\approx8.66
\end{aligned}
**第四步：softmax 得到 3 个概率**
\begin{aligned}
& p_{0}=\frac{7.39}{8.66}\approx0.853 \\
& p_{1}=\frac{1.22}{8.66}\approx0.141 \\
& p_{2}=\frac{0.05}{8.66}\approx0.006
\end{aligned}
#### **人话：**
> 对只复习了 1 小时的学生，模型认为：<br>
> 不及格：约 85.3% <br>
> 及格：约 14.1% <br>
> 优秀：约 0.6% <br>
> 模型基本在说：“八成多会挂科”，这很符合直觉。

### **4. 多分类交叉熵损失：公式 + 这个例子**
#### **4.1 真实标签怎么写？**
我们用 one-hot 表示真实类别，比如：
* 不及格（类别 0）--> y = [1, 0, 0]
* 及格（类别 1）--> y = [0, 1, 0]
* 优秀（类别 2）--> y = [0, 0, 1] <br>
对于学生 A（真实不及格）：
$$y^{(A)}=[1,0,0]$$

#### **4.2 单个样本的交叉熵损失公式**
对一个样本 i：
$$L^{(i)}=-\sum^2y_j^{(i)}\log p_j^{(i)}$$
因为 y 是 one-hot，只有真实类那一项是 1，其他是 0，所以实际上就是：
$$L^{(i)}=-\log(\text{模型给真实类别的概率})$$

#### **4.3 套到学生 A 上**
* 真实类别：不及格（0）
* 模型给不及格的概率：$p_0^{(A)}\approx0.853$
$$L^{(A)}=-\log(p_0)\approx-\log 0.853\approx0.159$$
#### **人话：**
> 模型认为“不及格”的概率是 85.3%，和真实情况（确实不及格）很接近，<br>
> 所以损失只有 0.159，挺小的 —— 表示这个样本上模型表现很好。

### **5. 学生 B：复习 3 小时，真实“及格”**
#### **5.1 计算打分**
\begin{aligned}
 & x=3 \\
 & z_{0}=-1\cdot3+3=0 \\
 & z_{1}=0.2\cdot3=0.6 \\
 & z_{2}=1\cdot3-4=-1
\end{aligned}

#### **5.2 取指数**
\begin{aligned}
& e^{z_0}=e^0=1.00 \\
& e^{z_1}=e^{0.6}\approx1.82 \\
& e^{z_2}=e^{-1}\approx0.37
\end{aligned}
求和：
\begin{aligned}
& S=1+1.82+0.37\approx3.19
\end{aligned}

#### **5.3 softmax 概率**
\begin{aligned}
& S=1+1.82+0.37\approx3.19 \\
& p_{0}=\frac{1}{3.19}\approx0.313 \
& p_{1}=\frac{1.82}{3.19}\approx0.571 \\
& p_{2}=\frac{0.37}{3.19}\approx0.116
\end{aligned}
人话：
> 对复习了 3 小时的学生，模型认为：
> * 不及格概率：31.3% <br>
> * 及格概率：57.1% <br>
> * 优秀概率：11.5% <br>
> 也就是说**模型最看好“及格”**，与真实情况（确实及格）一致，但还不够自信。

### **6. 学生 C：复习 6 小时，真实“优秀”**
#### **6.1 计算打分**
\begin{aligned}
 & x=6 \\
 & z_{0}=-1\cdot6+3=-3 \\
 & z_{1}=0.2\cdot6=1.2 \\
 & z_{2}=1\cdot6-4=2
\end{aligned}

#### **6.2 取指数**
\begin{aligned}
& e^{z_0}=e^{-3}\approx0.05 \\
& e^{z_1}=e^{1.2}\approx3.32 \\
& e^{z_2}=e^{2}\approx7.39
\end{aligned}
求和：
\begin{aligned}
& S=0.05+3.32+7.39\approx10.76
\end{aligned}

#### **6.3 softmax 概率**
\begin{aligned}
& p_{0}=\frac{0.05}{10.76}\approx0.005 \\
& p_{1}=\frac{3.32}{10.76}\approx0.309 \\
& p_{2}=\frac{7.39}{10.76}\approx0.686
\end{aligned}
人话：
> 对复习了 6 小时的学生，模型认为：
> * 不及格概率：0.5%（基本不可能）
> * 及格概率：30.9%
> * 优秀概率：68.6%（）最可能
> 也就是说，**模型看好“优秀”**。

#### **6.4 学生 C 的损失**
真实类别：优秀（2），用$p_2$：
$$L^{(C)}=-\log(p_2)\approx-\log 0.686\approx0.376$$
> 模型对“优秀”的信心是 68.7%，比 B 那个 57.1% 更强一些，<br>
> 所以损失 0.376 < 0.560，表现更好。

### **7. 如果模型严重看错，会怎样？**
假设有个学生 D：
* 复习 6 小时（x=6，打分和上面 C 一样）<br>
* 但这次**真实结果是“不及格”** 😱（可能临场发挥失常）<br>

那真实标签 one-hot 是：
$y^{(D)}=[1,0,0]$，我们要看的是$p_0\approx0.005$：
$$L^{(D)}=-\log(p_0)\approx-\log 0.005\approx5.38$$
> 这下损失就非常大了，<br>
> 因为模型说“他只 0.5% 可能不及格”，结果现实打脸：他就是不及格。
> 交叉熵就用一个**巨大的损失（5.38）**惩罚这个“非常自信但错得离谱”的预测。

这就是交叉熵的一个重要特点——**错得离谱，惩罚特别狠。**

### **8. 把多个学生的损失合起来**
假设我们有很多学生（样本 i），每个都有：
* 复习时间 $x^{(i)}$ <br>
* 模型算出来的概率 $p^{(i)}_0$，$p^{(i)}_1$，$p^{(i)}_2$ <br>
* 真实类别 one-hot 标签 $y^{(i)}=[y^{(i)}_0, y^{(i)}_1, y^{(i)}_2]$ <br>

那么**整个数据集的平均损失**是：
\begin{aligned}
J = \frac{1}{m}\sum_{i=1}^{m} L^{(i)}
  = -\frac{1}{m}\sum_{i=1}^{m}\sum_{j=0}^{2} y_{j}^{(i)} \log p_{j}^{(i)}
\end{aligned}
人话：
> 把每个学生上模型“对真实等级给的概率”的 log 损失加起来，再取平均。
> * 预测越准，真实等级概率越接近 1，损失就越小
> * 经常预测错，或者很自信地预测错，损失就会非常大

训练的过程就是：<br>
**不断调参数（这些线性函数里的 -1, 3, 0.2, -4 等），让这个平均损失 J 尽量小。**
