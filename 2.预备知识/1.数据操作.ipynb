```python
# 创建一个全0张量，形状为2x3x4
torch.zeros((2, 3, 4))
```

```python
# 创建一个全1张量，形状为2x3x4
torch.ones((2, 3, 4))
```

```python
# 为张量中的每个元素赋值
torch.tensor([[2, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
```

```python
# 常见的标准算数运算符都可以被升级为按元素运算
x = torch.tensor([1.0, 2, 4, 8])
y = torch.tensor([2, 2, 2, 2])
x + y, x - y, x * y, x / y, x ** y
```

```python
# 使用指数函数
torch.exp(x)
```

```python
# 连接多个张量
X = torch.arange(12, dtype=torch.float32).reshape((3, 4))
Y = torch.tensor([[2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
torch.cat((X, Y), dim=0), torch.cat((X, Y), dim=1)
```

```python
# 按元素比较两个张量
X == Y
```

```python
# 计算张量所有元素的和
X.sum()
```

```python
# 广播机制
a = torch.arange(3).reshape((3, 1))
b = torch.arange(2).reshape((1, 2))
a, b
```

```python
a + b # 广播机制使得不同形状的张量也能按元素运算，a和b分别被扩展为3x2的形状
```

```python
# 访问最后一行，以及第二行到第三行的元素
X, X[-1], X[1:3]
```

```python
# 通过索引来修改张量的元素
X[1, 2] = 9
X
```

```python
# 将前两行的所有元素设为12
X[0:2, :] = 12
X
```

```python
# 内存管理
before = id(Y)
Y = Y + X
id(Y) == before # False，因为Y被重新分配了内存
```

```python
# [:]操作符可以实现原地修改
Z = torch.zeros_like(Y) # 创建一个与Y形状相同的全0张量
print('id(Z):', id(Z))
Z[:] = X + Y
print('id(Z):', id(Z)) # id没有改变
```

```python
# +=运算符也是一个原地操作
before = id(X)
X += Y
id(X) == before # True，因为X的内存没有改变
```

```python
# 转换为NumPy张量
A = X.numpy()
B = torch.tensor(A)
type(A), type(B)
```

```python
# 将大小为1的张量转换为Python标量
a = torch.tensor([3.5])
a, a.item(), float(a), int(a)
```
