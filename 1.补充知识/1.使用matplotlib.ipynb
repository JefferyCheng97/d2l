```python
# 1. 简单的折线图
%matplotlib inline
import matplotlib.pyplot as plt

## 解决中文显示问题
plt.rcParams['font.sans-serif'] = ['SimHei']  # 黑体
plt.rcParams['axes.unicode_minus'] = False    # 正常显示负号
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

plt.plot(x, y)
plt.show()
```

```python
# 2. 自定义线条
plt.plot(x, y,
         color='red',
         linestyle='--',
         linewidth=2,
         marker='o',
         markersize=8)
plt.show()
```

```python
# 3. 设置标题、坐标轴标签、网格
plt.plot(x, y)

plt.title("折线图示例")
plt.xlabel("X 轴")
plt.ylabel("Y 轴")

plt.grid(True)
plt.show()
```

```python
# 4. 多线条
plt.plot([1,2,3], [1,2,3], label='Line 1')
plt.plot([1,2,3], [3,2,1], label='Line 2')

plt.legend() # 显示图例
plt.show()
```

```python
# 5. 面向对象的绘图
fig, ax = plt.subplots()

ax.plot(x, y)
ax.set_title("Title")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.show()
```

```python
# 6. 多子图
fig, axes = plt.subplots(1, 2, figsize=(10,4))

axes[0].plot([1,2,3], [1,4,2])
axes[0].set_title("Left")

axes[1].plot([1,2,3], [3,1,2])
axes[1].set_title("Right")

plt.show()
```

```python
# 7. 自定义图形大小和分辨率
plt.figure(figsize=(8, 4), dpi=120)
plt.plot(x, y)
plt.show()
```

```python
# 8. 保存图形
plt.savefig("./my_plot.png", dpi=200)
```
